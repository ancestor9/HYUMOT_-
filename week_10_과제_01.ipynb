```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
```

```python
X = df.drop('private', axis=1)
y = df['private']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

## <font color='red'> try and error
### ValueError: Invalid parameter weights for estimator Pipeline(steps=[('scaler', StandardScaler()), ('pca', PCA(n_components=0.8)),
                ('clf', KNeighborsClassifier())]). Check the list of available parameters with `estimator.get_params().keys()`.
를 수정하기

```python
from sklearn.model_selection import RandomizedSearchCV
param_dist= dict(n_neighbors=list(range(5,105, 2)),
                      weights=['uniform','distance'],
                      algorithm= ['ball_tree', 'kd_tree'],
                      leaf_size= list(range(5,50)))
```

```python
steps = [('scaler', StandardScaler()),
         ('pca', PCA(n_components=.80)),
         ('clf', KNeighborsClassifier())]

pipe_lr = Pipeline(steps)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, random_state=11)

rs = RandomizedSearchCV(pipe_lr, 
                        param_distributions= param_dist, cv=kf,
                        n_iter=64, scoring='recall_macro', verbose=True)
rs.fit(X_train, y_train)
```

```python
param_dist = {'clf__n_neighbors' : list(range(5,105, 2)),
              'clf__weights' : ['uniform','distance'],
              'clf__algorithm': ['ball_tree', 'kd_tree'],
              'clf__leaf_size' : list(range(5,50)),
              'clf__p' : [1, 2]
             }
```

```python
steps = [('scaler', StandardScaler()),
         ('pca', PCA(n_components=.80)),
         ('clf', KNeighborsClassifier())]

pipe_lr = Pipeline(steps)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, random_state=11)

rs = RandomizedSearchCV(pipe_lr, 
                        param_distributions= param_dist, cv=kf,
                        n_iter=64, scoring='recall_macro', verbose=True)
rs.fit(X_train, y_train)
```

```python
pipe_lr.fit(X_train, y_train)
train_predictions = pipe_lr.predict(X_train)
test_predictions = pipe_lr.predict(X_test)

print(classification_report(y_train, train_predictions))
```

```python
print(classification_report(y_test, test_predictions))
```
