```python
pytrend = TrendReq() 
kw_list = ['Salmonella', 'pesticide', 'BSE', 'Melamine', 'Norovirus'] 

df = pytrend.get_historical_interest(kw_list, year_start=2021, month_start=1, day_start=1, hour_start=0, 
                                     year_end=2021, month_end=7, day_end=23, hour_end=0, cat=0, 
                                     geo='', gprop='', sleep=0) 

colors = sns.color_palette("hls", len(kw_list)) 
for i, search_term in enumerate(kw_list): 
    df.plot(y=search_term, c=colors[i], figsize=(25, 3)) 
    plt.show()
```

```python
import sys
import pytrends 
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set() 
import pandas as pd 
from pytrends.request import TrendReq 
pytrend = TrendReq() 
kw_list = ['Salmonella', 'Norovirus'] 

df = pytrend.get_historical_interest(kw_list, year_start=2021, year_end=2021, 
                                     geo='', gprop='', sleep=0) 

colors = sns.color_palette("hls", len(kw_list)) 
for i, search_term in enumerate(kw_list): 
    df.plot(y=search_term, c=colors[i], figsize=(25, 3)) 
    plt.show()
```
