```python
import matplotlib.pyplot as plt 
# DATA PREPARATION
x = [1,2,3,4] 
y = [10,20,25,30]
# CREATE PLOT
fig = plt.figure() 
ax = fig.add_subplot(111) 
# PLOTTING
ax.plot(x, y, color='lightblue', linewidth=3) 
ax.scatter([2,4,6],  [5,15,25],  color='darkgreen',  marker='^')
# CUSTOMIZE PLOT
ax.set_xlim(1, 6.5)
# SAVE PLOT 
#plt.savefig('foo.png')
# SHOW PLOT 
plt.show()
```
