## Inferential statistics
Inferential statistical analysis infers properties of a population. for example by testing hypothesis and deriving estimates

```python
np.random.normal()
```

```python
population = np.random.normal(loc= 30, scale = 45, size=10000)  # 평균 30, 분산 45
population.shape
```

```python
from IPython.display import Image
Image('../image/boots.png', width=600, height=300)
```

```python
bootstrap =[]
estimates=[]

for i in range(1000):
    sample = np.random.choice(a=population, size=100)
    bootstrap.append(sample)
    estimates.append(sample.mean())
    
len(bootstrap)      # 200번 부트스트래핑
len(bootstrap[0])   # 재표본크기 100개
bootstrap[0][0:10]
```

```python
data = pd.DataFrame(bootstrap).T
data
```

```python
data.describe().head(3)
```

```python
data[0].plot(figsize=(20,5), kind='kde')
data[1].plot(figsize=(20,5), kind='kde')
data[2].plot(figsize=(20,5), kind='kde')
```

```python
oob = [x.round(2) for x in population if x not in data.values]
print(f'The Number of OOB(Out of Bags) Sample: {len(oob)}')
```
