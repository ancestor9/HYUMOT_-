## 사용자정의함수 만들기

```python
def my_square_sum(x):
    sum = 0
    for i in range(1,x):
        sum += i**2
    return sum
```

```python
my_square_sum(11) # 11(인자)를 입력하면 0부터 10까지의 수를 제곱합 총합을 결과 값으로 return
```

## <font color='red'> 분산과 공분산

```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
type(data)
data.keys()
```

```python
data.feature_names
data.target_names
```

```python
data.data.shape
data.data[0:10]
data.target[0:10]
data.target.shape
```

```python
features = data.data
```

### (1) features 공분산/상관관계 구하기

```python
np.cov(features.T)
```

```python
np.corrcoef(features.T)
```

### (2) mean centering한 features 공분산/상관관계 구하기

```python
np.mean(features, axis=0)
np.std(features, axis=0)
x_standard = (features - np.mean(features, axis=0)) / np.std(features, axis=0)
np.mean(x_standard, axis=0)
np.std(x_standard, axis=0)
```

```python
np.cov(x_standard.T)
```

```python
np.corrcoef(x_standard.T)
```

### (3) features와 target 상관관계 구하기

```python
np.corrcoef(features.T, data.target)
```
