# 문제 2. iris dataset의 공분산, 표준공분산

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston.feature_names
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['target'] = boston.target
df
```

####  자료 노이즈 만들기
1. 중복자료 생성

```python
df.head()
len(df)
df.sample(frac=0.01,random_state=430)
```

```python
df = df.append(df.sample(frac=0.01, random_state=1357))
df
```

2. Nan 생성('RM' 컬럼의 50개 행을 nan으로 무작위 생성)

```python
import numpy as np
np.random.seed(1357)
for i in np.random.choice(len(df), 30, replace=False):
    df.iloc[i, 5] = np.nan
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

### 문제 2.1  중복자료 row의 index를 기록하라

```python
df[df.duplicated()]
```

### 문제 2.2  중복자료를 제거한 후 index를 순번으로 재작성하라

```python
df[~df.duplicated()].reset_index().drop(['index'], axis=1, inplace = True)
df
```

### 문제 2.3 'RM'컬럼의 결측치를 median으로 대체하여 'RM'컬럼의 평균을 구하라

```python
df['RM'].isnull().sum()
```

```python
df['RM'].median()
```

```python
df['RM'] = df['RM'].fillna(df['RM'].median())
```

```python
df['RM'].mean()
```

# 문제 3. 자전거사고다발지의 3년치 데이터 EDA

```python
from glob import glob
glob('../data/*자전*')
```

```python
af = pd.read_csv('../data/2015년_자전거사고다발지.csv', encoding= 'euc-kr', usecols=[5,6])
bf = pd.read_csv('../data/2016년_자전거사고다발지.csv', encoding= 'euc-kr', usecols=[5,6])
cf = pd.read_csv('../data/2017년_자전거사고다발지.csv', encoding= 'euc-kr', usecols=[5,6])
```

### 3.1 파일별로 년도 컬럼을 생성하여 3개년도의 데이터를 모두 합쳐라

```python
af['년도'] = 2015; bf['년도'] = 2016; cf['년도'] = 2017
```

### 문제 3.1 년도별 자전거사고 발생건수가 가장 높은 년도?

```python
df = af.append(bf).append(cf)
df
```

```python
df.groupby('년도').sum()
```

### 문제 3.2 '시'와 '구'를 별도 컬럼으로 생성한 후 사고건수가 가장 높은 '시'를 확인하라

```python
df.head()
```

```python
df['시'] = df['다발지명'].str.split(' ', 2, expand=True)[0]
df['구'] = df['다발지명'].str.split(' ', 2, expand=True)[1]
df.head()
```

```python
df.groupby('시')['발생건수'].sum()
```

### 문제 3.3 사고건수가 가장 높은 '구'와 '년도'는 ?, 가장 낮은 '구'와 '년도'는 ?¶

```python
data = df.groupby(['시', '년도'])['발생건수'].sum().unstack().fillna(0)
data
```

```python
data.sum(1)
```

```python
data['ttl'] =data.sum(axis=1)
data.loc['ttl']=data.sum(0)
data.tail()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'
```
