```python
import pandas as pd
# 예제 데이터프레임 만들기

def make_df(var, obs):
    data = {c: [str(c) + str(i) for i in obs] 
            for c in var}
    return pd.DataFrame(data, obs)
```

```python
make_df('ABCD', [0,1,2,3]); make_df('012', ['A','B']); make_df('ACEGIJ', [1,3,5,7,9]) 
```

```python
df1 = make_df('ABCD', [0,1,2,3])
df2 = make_df('DEFG', [1,2,3,4])
```

```python
df1; df2; pd.concat([df1, df1])
```

# 1. Concat

```python
df1; df1; pd.concat([df1, df1]) # 기본은 상하로 합치기(vertical)
```

```python
df1; df1; pd.concat([df1, df1], axis=1) # axis=1, 좌우로 합치기(horizontal)
```

```python
df1; df2; pd.concat([df1, df2]) # 기본은 상하로 합치기, 합칠 떄 컬럼이 없으면 nan
```

```python
df1; df2; pd.concat([df1, df2], axis=0) # default는 axis=0
```

```python
df1; df2; pd.concat([df1, df2], ignore_index=True) # index를 새롭게 만들기
```

```python
df1; df2; pd.concat([df1, df2], axis=1) # 수평으로 합치기
```

```python
df1; df2; pd.concat([df1, df2], join='inner') # inner join(기본은 outer join), 서로 컬럼이 있는 경우만
```

```python
df1; df2; pd.concat([df1, df2], axis=1, join='inner') # 수평으로 서로 인덱스가 경우만
```

```python
df1; df2; pd.concat([df1.set_index('D'), df2.set_index('D')], axis=1) # 수평으로 특정 인덱스를 기준으로
```

# 2. append

### 왼쪽 파일을 기준으로(how='left') merge하기 때문에 왼쪽 파일의 D컬럼의 'D0'원소값을 포함한다

```python
df4; df5; pd.merge(df4, df5, how = 'left')
```

### merge할 경우 동일한 컬럼명인 경우 앞에 있는 컬럼에_x, 뒤의 컬럼에_y

```python
df6= df5[['A', 'B', 'D', 'H', 'I']]
df4; df6; pd.merge(df4, df6, on =['D', 'H'])
```

```python
df6= df5[['A', 'B', 'D', 'H', 'I']]
df4; df6; pd.merge(df4, df6, on =['D', 'H'], suffixes=('_left', '_right'))
```

```python
df1; df6; pd.merge(df1, df6, on=['D']) # 'D'를 기준으로 서로 합치기
```

```python
df1; df6; pd.merge(df1, df6, on=['A']) # 'A'를 기준으로 서로 합치기
```

```python
df1; df6; pd.merge(df1, df6, on=['A'], suffixes=('_left', '_right'))
```

### merge할 경우 컬럼명이 달라도 컬럼의 값이 동일하면 merge

```python
df6['J'] = df1['A']; df7 = df6[['H', 'J']]
df1; df7; pd.merge(df1, df7, left_on='A', right_on='J') # 컬럼명이 틀린 경우에도 서로 이름만 다를 경우
```
